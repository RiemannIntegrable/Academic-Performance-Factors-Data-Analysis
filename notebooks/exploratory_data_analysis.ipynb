## Treatment of Missing Data

```python
print("There are", external_data.shape[0] * external_data.shape[1], "data in total")
```

```python
print("There are", external_data.isnull().sum().sum(), "null data. It is the", round(external_data.isnull().sum().sum() / (external_data.shape[0] * external_data.shape[1]) * 100, 2), "% of total data")
```

```python
msno.matrix(external_data)
```
